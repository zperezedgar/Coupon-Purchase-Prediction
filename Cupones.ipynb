**Diseñamos una red neuronal con las características adecuadas para el conjunto de entrenamiento**

cuya salida es un softmax multiple (regresa un vector de probabilidades)

```
from keras import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU
from keras.layers import PReLU
from keras.layers import ELU
from keras.layers import Dropout
from keras.utils import plot_model

tamaño=df["COUPON_ID_hash"].size #dimension de los vectores a crear
#tamaño=12

def build_regressor():
    regressor = Sequential()
    regressor.add(Dense(units=2, input_shape = (2,)))
    regressor.add(ELU(alpha=1)) 
    regressor.add(Dense(units=int(round((tamaño)/3))))
    regressor.add(ELU(alpha=1))
    regressor.add(Dropout(0.25)) #dropout de 25% (neuronas a apagar en cada epoch)
    regressor.add(Dense(units=tamaño,activation='softmax')) 
    regressor.compile(optimizer='Nadam', loss='categorical_crossentropy',  metrics=['accuracy'])
    return regressor

build_regressor().summary()
```

**entrenamos la red neuronal con el conjunto de entrenamiento X y el vector target T**

```
iteraciones=20 #epochs
from keras.wrappers.scikit_learn import KerasRegressor
regressor = KerasRegressor(build_fn=build_regressor, batch_size=20,epochs=iteraciones,validation_split=0.2, shuffle=True,verbose=2)
```

```
import time #para ver cuanto tarda
tic=time.time()

results=regressor.fit(np.array(X),np.array(T))
print('Tiempo de procesamiento (secs): ', time.time()-tic)
```

Nuestro modelo mide la afinidad de cada usuario por los cupones. Una alta afinidad por algún cupón indica que es más probable que el usuario compre algún boleto.

```
import random

#NUEVO USUARIO
age=random.randint(1,100)
sex=np.random.choice(2) #0=hombre, 1=mujer
if sex==0:
  sex_str="M"
else:
  sex_str="F"

#predecimos las probabilidades de compra
#len(diccionario_cupones) #número de cupones existentes
arr=regressor.predict(np.array([[age,sex]]))
sum=0
for i in range(1,11): #sumamos las probabilidades de los 100 cupones mas probables de ser vendidos
  sum+=np.partition(arr.flatten(), -2)[-i]
if sum>0.1:
  guarantee="YES"
else:
  guarantee="NO"

#buscamos el cupón mas probable de comprar
max_prob=np.amax(arr)
result = np.where(arr == np.amax(arr))
tamaño=df["COUPON_ID_hash"].size #dimension de los vectores
vector=np.zeros(tamaño)
vector[result[1][0]]=1

 
#print('Max prob : ', max_prob)
#print('Returned tuple of arrays :', result)
#print('Index of maximum element :', result[1])
print("AGE: ", age)
print("SEX: ", sex_str)
print("Coupon most probable to be sold: ", [k for k,v in diccionario_cupones.items() if np.array_equal(v,vector)])
print('Prob. of selling the coupon most probable to be sold: ', round(max_prob*100,3),"%")
print('Prob. of selling any of the 10 coupons most probable to be sold: ', round(sum*100,3),"%")
print("Almost Ensured sell?", guarantee) #medimos afinidad por los cupones hacia el usuario
```
